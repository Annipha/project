# Data Visualization for household attributes

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("final raw data.csv")
data.head(5)
data.rename(columns={'Full count Larvae': 'Larvae_count'}, inplace=True)
```

# Data Visualization

```python
0# Larvae count per hamlet
%matplotlib inline
import matplotlib.pyplot as plt
hamlet_larvae = data[data['Larvae'] == 1]
plt.scatter(data.Hamlet, data.Larvae_count)
plt.xlabel("Hamlet")
plt.ylabel("Larvae count")
plt.title("Hamlet vs. Larvae count")
plt.ylim(0, 350)
plt.xlim(-1, 12)
```

```python
# count of larvae per hamlet
hamlet_larvae = data[data['Larvae'] == 1]
new_hamlet_larvae = hamlet_larvae[['Hamlet','Larvae_count']]
new_hamlet_larvae.groupby('Hamlet').mean().plot(kind='bar')
plt.xlabel("Hamlet")
plt.ylabel("Average count of larvae")
plt.title("Average number of larvae per hamlet")
```

```python
# households per a hamlet with larvae
hamlet_larvae = data[data['Larvae'] == 1]
new_hamlet_larvae = hamlet_larvae[['Hamlet','Larvae_count']]
new_hamlet_larvae.Hamlet.value_counts().plot(kind='bar')
plt.xlabel("Hamlet")
plt.ylabel("Households with larvae")
plt.title("Households with mosquito larvae per hamlet")
plt.ylim(0, 10)
```

```python
#group larvae by no. of containers
```

```python
from scipy.stats import gaussian_kde
vstacked = np.vstack([hamlet_larvae.Hamlet, hamlet_larvae.Larvae_count])
density_color = gaussian_kde(vstacked)(vstacked)
idx = density_color.argsort()
hamlet_larvae.Hamlet = hamlet_larvae.Hamlet[idx]
hamlet_larvae.Larvae_count = hamlet_larvae.Larvae_count[idx]
density_color = density_color[idx]
```

```python
fig, ax = plt.subplots()
ax.scatter(hamlet_larvae.Hamlet, hamlet_larvae.Larvae_count, c=density_color, s=50, edgecolor='')
plt.xlim(-2, 12)
plt.ylim(1, 300)
plt.xlabel("Hamlet")
plt.ylabel("Count of larvae")
plt.title("Density plot of count of larvae in a hamlet")
plt.show()
```
