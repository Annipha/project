```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

```python
data = pd.read_csv('STATA.csv', na_filter=False)

data.head()
```

```python
#data.plot(kind='scatter',x='Water_level',y='Full count Larvae')
data.shape
annplot=data[['Water_level','count_larvae']]
annplot.shape
annplot.head()
```

```python
annplot.count_larvae.plot(kind = 'hist', bins = 20, range=[1, 100])
plt.ylim((0,10))
plt.xlabel('Mosquito laevae count')
plt.ylabel('Frequency')
```

```python

x = annplot.plot(kind='scatter',x='Water_level',y='count_larvae')
x.set_ylim(10,300)
```

```python
a = data.plot(kind = 'scatter', x = 'Water_level', y ='count_larvae', c='Full count_pupae',colormap = 'Greens')
a.set_ylim(10,300)
```

```python
pd.scatter_matrix(data[['count_larvae','Water_level','Full count_pupae']],figsize=(10,8))
```
