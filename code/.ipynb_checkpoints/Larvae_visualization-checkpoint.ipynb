```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

```python
data = pd.read_csv('STATA.csv', na_filter=False)

data.head()
```

```python
#data.plot(kind='scatter',x='Water_level',y='Full count Larvae')
data.shape
annplot=data[['Water_level','Full count Larvae']]
annplot.shape
annplot.head()
```

```python
annplot.Laevae_count.plot(kind = 'hist', bins = 20, range=[1, 100])
plt.ylim((0,10))
plt.xlabel('Mosquito laevae count')
plt.ylabel('Frequency')
```

```python
#drinks.plot(kind='scatter',x='beer',y='wine')
annplot.plot(kind='scatter',x='Water_level',y='Full count Larvae')
#data.head()
```

```python
data.plot(kind = 'scatter', x = 'Water_level', y ='Full count Larvae', c='Full count Pupae',colormap = 'Greens')
```

```python
pd.scatter_matrix(data[['Full count Larvae','Water_level','Full count Pupae']],figsize=(10,8))
```
